```python
x= range(0,11)
y= range(5,16)
plt.plot(x,y,'m^')
#. ''안에 b g r c청록 m보라 y노르스름 k검정 w
#. o -> 원, v -> 역삼각형, ^-> 삼각형
#. s -> 사각형, + -> +, . -> .
#. ''안에 ro -> 빨간색 원
#. ''안에 m^ -> 보라색 삼각형
plt.show()
```

```python
#그림을 그리기 위해 도화지 준비
fig = plt.figure()

# 큰도화지 안에 작은 구역으로 나누기
# 행의갯수, 열의 갯수, 작성될 위치
a1 = fig.add_subplot(1,2,1)
x= range(0,11)
y= range(5,16)
a1.plot(x,y,'m^')
a1.set_xlabel('time')
a1.set_ylabel('money')


a2 = fig.add_subplot(2,2,2)
x= range(0,50)
y= range(0,50)
a2.plot(x,y,'m^')
a2.set_xlabel('time')
a2.set_ylabel('money')


fig.add_subplot(2,2,4)
plt.show()
```

```python
from tkinter import font
import matplotlib
import matplotlib.font_manager as fm

fig = plt.figure(constrained_layout=True)

font_location = "C:\Windows\Fonts\H2PORM.TTF"
fontname = fm.FontProperties(fname = font_location).get_name()
matplotlib.rc('font',family = fontname)

plt.style.use('ggplot')
a= fig.add_subplot(111)
x=[3,4]
y=[1,2]
labels = ['english', 'math']
colors = ['green', 'purple']


# 숫자 어느 그래프에, 그래프와 숫자와 거리
a.bar(x,y, tick_label = labels, color = colors ) #barh -> 가로로
for i, v in enumerate(y) : 
    a.text(v+2, i+0.5, str(v))
plt.show()
```

```python
#데이터 프레임 생성
#딕셔러니와 리스트 활용
#이름 -> 집 -> 나이
#이름 -> 나이 -> 집 순서 변경하기
member1 ={'이름' : ['홍길동', '홍길순','임꺽정'], '나이' : ['23','22','50'], '집' : ['대전','서울','부산']}
member2 = pd.DataFrame(member1, columns =['이름','집','나이'])
member2
```

```python

name =  ['홍길동', '홍길순','임꺽정']
age =  [23,22,50]
home = ['대전','서울','부산']
member3 = pd.DataFrame()
member3['이름'] = name
member3['나이'] = age
member3['집'] = home

# sex = ['남자','여자','남자']
# member3['성별'] = sex
gender = pd.Series(['남자', '여자', '남자'] , name='성별')
member3['성별'] = gender
print(member3)

print(member3['이름'])
    # 특정 열(column) 만 보이게
print(member3['나이'].min())
print(member3['나이'].max())
print(member3['나이'].mean())
print(member3.describe())
print(member3.head(2))
print(member3.tail(1))
```

```python
# 값이 큰것 부터!! 정렬하기 -> .sort_values(['index이름'], ascending=[False])

number = pd.Series([1, 2, 3] , name='번호') 
name= pd.Series(['전우치', '강감찬','홍길동'] , name='이름')
birthDay = pd.Series([1999, 2000, 2001] , name='생년')
member4 = pd.DataFrame()
member4['번호'] = number
member4['이름'] = name
member4['생년'] = birthDay
print(member4)
print(member4.sort_values(['생년'], ascending=[False]))

```

```python
import numpy as np
x = [1, 2]
y1, y2, y3 = [1,2], [2,5], [3,7]
fig = plt.figure()
ax = fig.add_subplot(111)
w = 0.2
ax.bar(x,y1,width=w) 
ax.bar(np.array(x)      ,y1 ,width=w ,   label='english')
ax.bar(np.array(x)+ w   ,y2 ,width=w ,   label='math')
ax.bar(np.array(x)+ w*2 ,y3 ,width=w ,   label='korean')
ax.legend()
plt.show()   
```
