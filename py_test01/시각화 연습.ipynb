```python
#시각화 모듈
import matplotlib.pyplot as plt
plt.style.use('default')
plt.plot([2,4,6,8,1,3,5,7,9])
plt.show()
```

```python
x=[2,4,6,8]
y=[1,3,5,7]
plt.plot(x,y,'r')
plt.show()
```

```python
x= range(0,11)
y= range(5,16)
plt.plot(x,y,'m^')
#. ''안에 b g r c청록 m보라 y노르스름 k검정 w
#. o -> 원, v -> 역삼각형, ^-> 삼각형
#. s -> 사각형, + -> +, . -> .
#. ''안에 ro -> 빨간색 원
#. ''안에 m^ -> 보라색 삼각형
plt.show()
```

```python
#그림을 그리기 위해 도화지 준비
fig = plt.figure()

# 큰도화지 안에 작은 구역으로 나누기
# 행의갯수, 열의 갯수, 작성될 위치
a1 = fig.add_subplot(1,2,1)
x= range(0,11)
y= range(5,16)
a1.plot(x,y,'m^')
a1.set_xlabel('time')
a1.set_ylabel('money')


a2 = fig.add_subplot(2,2,2)
x= range(0,50)
y= range(0,50)
a2.plot(x,y,'m^')
a2.set_xlabel('time')
a2.set_ylabel('money')


fig.add_subplot(2,2,4)
plt.show()
```

```python
fig = plt.figure()
plt.style.use('ggplot')
a= fig.add_subplot(111)
x=[3,4]
y=[1,2]
labels = ['english', 'math']
colors = ['green', 'purple']

a.bar(x,y, tick_label = labels, color = colors )

plt.show()
```
